```python
import pandas as pd
data = pd.read_csv('insurance.csv')
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data[data.duplicated()]
```

```python
data.drop_duplicates()
```

```python
data = data.drop_duplicates().reset_index(drop=True)
data
```

```python
X = data.drop('charges',axis=1)
```

```python
X
```

```python
y = data['charges']
```

```python
y
```

```python
import numpy as np
X.sex = np.where(X.sex == 'female',0,1)
```

```python
X.head()
```

```python
X.smoker = np.where(X.smoker == 'no',0,1)
```

```python
X.head()
```

```python
X.region.unique()
```

```python
X = pd.get_dummies (X, columns= ['region'])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=123)
X_train.shape
```

```python
#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#scaler.fit_transform(X_train,y_train)
#scaler.transform(X_test,y_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann = Sequential()
```

```python
ann.add(Dense(units = 9,input_dim = X_train.shape[1],kernel_initializer = 'random_uniform',activation = 'relu'))
```

```python
ann.add(Dense(units = 5,kernel_initializer = 'random_uniform',activation = 'relu'))
```

```python
ann.add(Dense(units=1,kernel_initializer='random_uniform',activation = 'relu'))
```

```python
ann.compile(loss = 'mean_squared_error',optimizer = 'adam')
```

```python
ann.summary()
```

```python
ann.fit(X_train,y_train,batch_size=15,epochs=500,validation_data=(X_test,y_test))
```

```python
y_pred = ann.predict(X_test)
```
